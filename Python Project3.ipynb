##   Data Visualization

```python
#Visualize Distributions
plt.figure(figsize=(10, 6))
sns.barplot(x='co2', y='VClass', data=df, ci=None)
plt.title('Average CO2 Emissions by Vehicle Class')
plt.xlabel('Average CO2 Emissions (g/km)')
plt.ylabel('Vehicle Class')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='displ', y='co2', hue='fuelType', data=df)
plt.title('CO2 Emissions vs. Engine Displacement')
plt.xlabel('Engine Displacement (Liters)')
plt.ylabel('CO2 Emissions (g/km)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='cylinders', y='comb', data=df)
plt.title('Fuel Efficiency by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Combined Fuel Efficiency (MPG)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_grouped = df.groupby('year')['comb'].mean().reset_index()
sns.lineplot(x='year', y='comb', data=df_grouped)
plt.title('Average Fuel Efficiency Over Time')
plt.xlabel('Year')
plt.ylabel('Average Combined Fuel Efficiency (MPG)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['ghgScore'], bins=10, kde=True)
plt.title('Distribution of Greenhouse Gas Score')
plt.xlabel('Greenhouse Gas Score')
plt.ylabel('Frequency')
plt.show()
```

```python
fuel_type_counts = df['fuelType'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(fuel_type_counts, labels=fuel_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Fuel Types')
plt.show()
```

```python
vclass_counts = df['VClass'].value_counts()

plt.figure(figsize=(10, 8))
plt.pie(vclass_counts, labels=vclass_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Vehicle Classes')
plt.show()
```
